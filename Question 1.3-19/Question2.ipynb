```python
import os
import time
from random import *
import random


class SMS:
    # Menu
    def menu(self):
        print("1. Admin")
        print("2. Student")
        c = input("Choose from above: ")
        if c == "1":
            SMS.admin(self)
        elif c == "2":
            SMS.student(self)
        else:
            print("Invalid choice")
            SMS.menu(self) 

    def student(self):
        print("Student account")
        print("---------------")
        print("1. Login")
        print("2. Register")
        c = input("Choose from above: ")
        if c == "1":
            SMS.studentLoginn(self)
        elif c == "2":
            SMS.studentRegister(self)
        else:
            print("Invalid choice")
            SMS.student(self)

    def studentLoginn(self):
        print("Login as a student")
        print("------------------")
        fname = input("Firstname: ")
        ID__ = input("ID: ")
        for i in os.listdir("C:"):
            try:
                if str(fname)+"."+str(ID__) in i:
                    SMS.studentDashboard(self)
            except:
                print("Invalid details!")
                SMS.studentLogin(self)

    ############################################ Student registration ##############################
    def studentRegister(self):
        print("Student registration")
        print("--------------------")
        fname = input("Firstname: ")
        lname = input("Lastname: ")
        age = input("Age: ")
        gender = input("Gender: ")
        ID = randint(1000000, 9000000)
        if fname != "" or lname != "" or age != "" or gender != "":
            with open(str(fname)+"."+str(ID), "a") as f:
                f.write("Firstname: "+str(fname)+"\n")
                f.write("Lastname: "+str(lname)+"\n")
                f.write("Age: "+str(age)+"\n")
                f.write("Gender: "+str(gender)+"\n")
                f.write("ID: "+str(ID)+"\n")
                f.write("\n")
                SMS.studentDashboard(self)
        else:
            print("All fields are required!")
            SMS.studentRegister(self)

##############################################Student login ####################################
    def studentLogin(self):
        print("Student Login")
        print("-------------")
        name = input("Firstname: ")
        ID = input("ID: ")
        if name == "" or ID == "":
            print("All fields are required!")
        else:
            o = os.listdir("C:")
            for reg in o:
                if str(name)+"."+str(ID) not in reg:
                    SMS.studentDashboard(self)
    
                else:
                    print("Account is not registered!")
                    SMS.studentLogin(self)
     

    def studentDashboard(self):
        print("")
        print("Student dashboard")
        print("---------------")
        print("1. My details")
        print("2. Messages")
        print("3. Logout")
        c = input("Choose from above: ")
        if c == "1":
            try:
                name = input("Firstname: ")
                ID = input("ID: ")
                with open(str(name)+"."+str(ID), "r") as f:
                    print(f.read())
                    print("")
                    SMS.studentDashboard(self)
            except:
                print("Student could not be found!")
                SMS.studentDashboard(self)
        elif c == "2":
            SMS.studentMessages(self)
        elif c == "3":
            SMS.student(self)
        else:
            print("Invalid entry!")

    def studentMessages(self):
        print("Messages")
        print("1. Send message")
        print("2. View message")
        c = input("Choose from above: ")
        if c == "1":
            m = input("Message:-->: ")
            fname = input("Firstname: ")
            ID = input("ID: ")
            for i in os.listdir("C:"):
                if str(fname)+"."+str(ID) not in i:
                    print("Invalid details")
                else:
                    with open(str(fname)+".message", "w") as mess:
                        mess.write(m+"\n")
        elif c == "2":
            ID_ = input("ID: ")
            for x in os.listdir("C:"):
                if str(ID_)+"."+"adminMessage" not in x:
                    print("No message for you!")
                    break
                else:
                    try:
                        with open(str(ID_)+"."+"adminMessage", "r") as fi:
                            print(fi.read())
                    except:
                        print("No message for you!")
                        SMS.studentMessages(self)
                        break
        else:
            print("Invalid choice!")
            SMS.studentMessages(self)


    def admin(self):
        print("Admin account")
        print("=============")
        name = input("Username: ")
        password = input("Password: ")
        if name == "Daniel" and password == "5716":
            SMS.adminDashboard(self)

        else:
            print("Invalid admin details")


    def adminDashboard(self):
        print("")
        print("Admin dashboard")
        print("===============")
        print("1. View Student's details")
        print("2. Messages")
        print("3. Remove student")
        print("4. Sign in a new student")
        print("5. Logout")
        c = input("Choose from above: ")
        if c == "1":
            try:
                name = input("Firstname: ")
                IDDD = input("ID: ")
                with open(str(name)+"."+str(IDDD), "r") as f:
                    print(f.read())
                    SMS.adminDashboard(self)
            except:
                print("Student could not be found!")
                SMS.adminDashboard(self)
        elif c == "2":
            SMS.adminMessages(self)

        elif c == "3":
            fname = input("Firstname: ")
            ID_ = input("ID: ")
            for x in os.listdir("C:"):
                try:
                    if str(fname)+"."+str(ID_) in x:
                        os.remove(str(fname)+"."+str(ID_))
                        print("Deleting account...")
                        time.sleep(2)
                        print("Account deleted successfully!")
                        SMS.admin(self)
                except:
                    print("Account could not be found!")
        elif c == "4":
            print("Sign in a new student")
            print("=====================")
            fname = input("Firstname: ")
            lname = input("Lastname: ")
            age = input("Age: ")
            gender = input("Gender: ")
            ID = randint(1000000, 9000000)
            if fname != "" or lname != "" or age != "" or gender != "":
                with open(str(fname)+"."+str(ID), "a") as f:
                    f.write("Firstname: "+str(fname)+"\n")
                    f.write("Lastname: "+str(lname)+"\n")
                    f.write("Age: "+str(age)+"\n")
                    f.write("Gender: "+str(gender)+"\n")
                    f.write("ID: "+str(ID)+"\n")
                    f.write("\n")
                    print("Success! ")
                    SMS.adminDashboard(self)
            else:
                print("All fields are required!")
                SMS.adminDashboard(self)

        elif c == "5":
            SMS.admin(self)


    def adminMessages(self):
        print("Messages")
        print("========")
        print("1. View message")
        print("2. Send message")
        c = input("Choose from above: ")
        if c == "1":
            n = input("Firstname: ")
            for i in os.listdir("C:"):
                if str(n)+".message" not in i:
                    print("No messages to view!")
                    SMS.adminMessages(self)
                else:
                    try:
                        with open(str(n)+".message", "r") as f:
                            print(f.read())
                    except:
                        print("No message to view!")
                        SMS.adminMessages(self)
        elif c == "2":
            message = input("Message:-->: ")
            _ID = input("Student ID: ")
            with open(str(_ID)+".adminMessage", "w") as adminM:
                adminM.write(str(message)+"\n")
                adminM.close()
                SMS.adminMessages(self)
                SMS.menu(self)
```
