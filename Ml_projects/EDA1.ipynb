```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("lung_cancer.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
### No missing values
## Unique values
for i in df.columns:
    print("Unique categories in",i,"is: ", df[i].unique())
    
## 2 means Yes and 1 means No
## Simlarly, let 2 means M and 1 means F
```

```python
df['GENDER'] = df['GENDER'].map({"M": 2, "F": 1})
df['LUNG_CANCER'] = df['LUNG_CANCER'].map({"NO": 1, "YES": 2})
```

```python
df.dtypes
```

```python
### checking for duplicates:
df.duplicated().sum()

df.drop_duplicates(inplace= True)
```

```python
categorical_data = df.drop(columns= ['AGE', 'LUNG_CANCER'])
for i in categorical_data.columns:
    plt.figure(figsize = (9,6))
    plt.title(i + ' Distribution', weight = 'bold')
    sns.countplot(x = categorical_data[i], palette = 'Set1')
    plt.show()
    print()
```

```python
for i in categorical_data.columns:
    print("Count of",i,"is: ")
    print(df.groupby(i)["LUNG_CANCER"].count())
    print("-------------------------------------------------")
```

```python
for i in categorical_data.columns:
    plt.figure(figsize = (9,6))
    plt.title(i + '-LUNG CANCER Relationship', weight = 'bold')
    sns.countplot(x = categorical_data[i], hue = df['LUNG_CANCER'], palette = 'Set1')
    plt.show()
    print()
    
    
    ## From here we can clearly understand how lung cancer varies on various factors
    ## 2 means yes and 1 means no
    ## for gender 2 male and 1 female
```

```python
### Insights are lung cancer is unexpected as it depends on all factors above (50 - 50 perceent) ## it may depend on geographical facrtors, lifestyle etc
```
